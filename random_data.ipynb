```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt

def random_line(m, sigma, size=10):
    xdata = np.linspace(0.0,2.0,size)
    # Generate normally distributed random error ~ N(0, sigma**2)
    errors = scipy.stats.norm.rvs(loc=0, scale=sigma, size=size)
    ydata = m * xdata  + errors
    return xdata, ydata

xs, ys = random_line(2,  1, size=100)

# Plot to see how closely the values fit the 
#   original line
fig, ax = plt.subplots()
ax.plot(xs, ys, 'o')
ax.plot(xs, 2 * xs )
plt.xlabel(r'$M$')
plt.ylabel(r'$r_s$')
plt.title('Adding a Gaussian noise to ' r'$r_s$ ' 'with ' r'$\theta =1$')
plt.show()
```
